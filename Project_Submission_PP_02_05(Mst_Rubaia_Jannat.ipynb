# **1.Event Create Management**

```python
import csv
import os
from datetime import datetime, timedelta
import calendar
#1.Event Create and Management
# File paths
EVENT_FILE = "events.csv"
ATTENDEE_FILE = "attendees.csv"
BUDGET_FILE = "budgets.csv"

# Ensure the events file exists and contains headers
if not os.path.exists(EVENT_FILE):
    with open(EVENT_FILE, 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(["Event Name", "Date and Time", "Location", "Description"])
#1.1(Event create and Management)
# Function to create an event
def create_event():
    name = input("Enter Event Name: ")
    date_time = input("Enter Date and Time (YYYY-MM-DD HH:MM): ")
    location = input("Enter Location: ")
    description = input("Enter Description: ")

    with open(EVENT_FILE, 'a', newline='') as file:
        writer = csv.writer(file)
        writer.writerow([name, date_time, location, description])
#1.2(Event create and management)
# Function to edit an event
def edit_event():
    view_events()
    event_index = int(input("Enter the number of the event to edit: ")) - 1

    with open(EVENT_FILE, 'r') as file:
        reader = csv.reader(file)
        events = list(reader)

    if 0 <= event_index < len(events) - 1:
        event = events[event_index]
        name = input(f"Enter new Event Name (current: {event[0]}): ") or event[0]
        date_time = input(f"Enter new Date and Time (current: {event[1]}): ") or event[1]
        location = input(f"Enter new Location (current: {event[2]}): ") or event[2]
        description = input(f"Enter new Description (current: {event[3]}): ") or event[3]

        events[event_index] = [name, date_time, location, description]

        with open(EVENT_FILE, 'w', newline='') as file:
            writer = csv.writer(file)
            writer.writerows(events)

        print("Event updated successfully!")
    else:
        print("Invalid event number.")
#1.3(Event Create and Management)
# Function to delete an event
def delete_event():
    view_events()
    event_index = int(input("Enter the number of the event to delete: ")) - 1

    with open(EVENT_FILE, 'r') as file:
        reader = csv.reader(file)
        events = list(reader)

    if 0 <= event_index < len(events) - 1:
        deleted_event = events.pop(event_index)

        with open(EVENT_FILE, 'w', newline='') as file:
            writer = csv.writer(file)
            writer.writerows(events)

        print(f"Deleted event: {deleted_event[0]}")
    else:
        print("Invalid event number.")
# Helper function to view events
def view_events():
    with open(EVENT_FILE, 'r') as file:
        reader = csv.reader(file)
        events = list(reader)
    if len(events) <= 1:
        print("No events found.")
    else:
        for i, event in enumerate(events[1:], start=1):
            print(f"{i}. {event[0]} | {event[1]} | {event[2]} | {event[3]}")
```

# **2.Attendees Management**

```python
#2.Ateendees Management
# Ensure the attendees file exists and contains headers
if not os.path.exists(ATTENDEE_FILE):
    with open(ATTENDEE_FILE, 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(["Event Name", "Attendee Name", "Attendee Email", "RSVP Status"])
#2.1(Attendee Management)
# Function to add an attendee
def add_attendee():
    event_name = input("Enter the Event Name for the attendee: ")
    attendee_name = input("Enter Attendee Name: ")
    attendee_email = input("Enter Attendee Email: ")
    rsvp_status = input("Enter RSVP Status (Confirmed/Declined/Maybe): ")

    with open(ATTENDEE_FILE, 'a', newline='') as file:
        writer = csv.writer(file)
        writer.writerow([event_name, attendee_name, attendee_email, rsvp_status])

    print(f"Attendee {attendee_name} added to {event_name}.")
#2.2
# Function to view attendees
def view_attendees():
    event_name = input("Enter the Event Name to view attendees: ")

    with open(ATTENDEE_FILE, 'r') as file:
        reader = csv.reader(file)
        attendees = list(reader)

    event_attendees = [attendee for attendee in attendees[1:] if attendee[0].lower() == event_name.lower()]

    if event_attendees:
        print(f"Attendees for {event_name}:")
        for attendee in event_attendees:
            print(f"{attendee[1]} ({attendee[2]}) - RSVP Status: {attendee[3]}")
    else:
        print(f"No attendees found for {event_name}.")
#2.3
# Function to remove an attendee
def remove_attendee():
    event_name = input("Enter the Event Name to remove attendee from: ")
    attendee_name = input("Enter Attendee Name to remove: ")

    with open(ATTENDEE_FILE, 'r') as file:
        reader = csv.reader(file)
        attendees = list(reader)

    updated_attendees = [attendee for attendee in attendees if not (attendee[0].lower() == event_name.lower() and attendee[1].lower() == attendee_name.lower())]

    if len(updated_attendees) < len(attendees):
        with open(ATTENDEE_FILE, 'w', newline='') as file:
            writer = csv.writer(file)
            writer.writerows(updated_attendees)
        print(f"Removed {attendee_name} from {event_name}.")
    else:
        print(f"Attendee {attendee_name} not found in {event_name}.")
```

# **3.Schedule Management**

```python
#3.SChedule Management
# Function to display upcoming events
def display_upcoming_events():
    now = datetime.now()
    with open(EVENT_FILE, 'r') as file:
        reader = csv.reader(file)
        events = list(reader)

    upcoming_events = [event for event in events[1:] if datetime.strptime(event[1], "%Y-%m-%d %H:%M") > now]

    if upcoming_events:
        print("Upcoming Events:")
        for event in upcoming_events:
            print(f"{event[0]} | {event[1]} | {event[2]} | {event[3]}")
    else:
        print("No upcoming events.")
#3.1
# Function to filter events by date or location
def filter_events():
    filter_type = input("Filter by (1) Date or (2) Location: ")

    with open(EVENT_FILE, 'r') as file:
        reader = csv.reader(file)
        events = list(reader)

    if filter_type == '1':
        date_str = input("Enter date (YYYY-MM-DD): ")
        date_obj = datetime.strptime(date_str, "%Y-%m-%d")
        filtered_events = [event for event in events[1:] if datetime.strptime(event[1], "%Y-%m-%d %H:%M").date() == date_obj.date()]
    elif filter_type == '2':
        location = input("Enter location: ")
        filtered_events = [event for event in events[1:] if location.lower() in event[2].lower()]
    else:
        print("Invalid filter type.")
        return

    if filtered_events:
        for event in filtered_events:
            print(f"{event[0]} | {event[1]} | {event[2]} | {event[3]}")
    else:
        print("No events found matching the filter.")
#3.2
# Function to send reminder for events within 24 hours
def send_event_reminders():
    now = datetime.now()
    reminder_time = now + timedelta(hours=24)

    with open(EVENT_FILE, 'r') as file:
        reader = csv.reader(file)
        events = list(reader)

    reminders = [event for event in events[1:] if now <= datetime.strptime(event[1], "%Y-%m-%d %H:%M") <= reminder_time]

    if reminders:
        for event in reminders:
            print(f"Reminder: {event[0]} is scheduled for {event[1]} at {event[2]}")
    else:
        print("No events in the next 24 hours.")
```

# **4.Budget Management**

```python
#4.1(Budget Management)
# Function to set or update event budget
def set_budget():
    event_name = input("Enter the Event Name to set budget for: ")
    total_budget = float(input("Enter Total Budget: "))
    venue = float(input("Enter Venue Budget: "))
    catering = float(input("Enter Catering Budget: "))
    entertainment = float(input("Enter Entertainment Budget: "))
    miscellaneous = float(input("Enter Miscellaneous Budget: "))

    with open(BUDGET_FILE, 'a', newline='') as file:
        writer = csv.writer(file)
        writer.writerow([event_name, total_budget, venue, catering, entertainment, miscellaneous])

    print(f"Budget set for {event_name}.")
#4.2
# Function to view budget status for an event
def view_budget():
    event_name = input("Enter the Event Name to view budget: ")

    with open(BUDGET_FILE, 'r') as file:
        reader = csv.reader(file)
        budgets = list(reader)

    event_budget = next((budget for budget in budgets[1:] if budget[0].lower() == event_name.lower()), None)

    if event_budget:
        total_budget = float(event_budget[1])  # Convert to float
        expenses = {
            "Venue": float(event_budget[2]),  # Convert to float
            "Catering": float(event_budget[3]),  # Convert to float
            "Entertainment": float(event_budget[4]),  # Convert to float
            "Miscellaneous": float(event_budget[5])  # Convert to float
        }
        total_expenses = sum(expenses.values())
        remaining_budget = total_budget - total_expenses

        print(f"Budget for {event_name}:")
        for category, amount in expenses.items():
            print(f"{category}: ${amount:.2f}")
        print(f"Total Expenses: ${total_expenses:.2f}")
        print(f"Remaining Budget: ${remaining_budget:.2f}")
        if remaining_budget < 0:
            print("Warning: Budget exceeded!")
    else:
        print(f"No budget information found for {event_name}.")
```

# **Additional Features**

```python
#Additional Features (send Invitations)
def search_events():
    """Searches events based on user input."""
    search_term = input("Enter search term (event name, location, etc.): ")

    with open(EVENT_FILE, 'r') as file:
        reader = csv.reader(file)
        events = list(reader)

    matching_events = [event for event in events[1:] if search_term.lower() in " ".join(event).lower()]

    if matching_events:
        print("Matching Events:")
        for event in matching_events:
            print(f"{event[0]} | {event[1]} | {event[2]} | {event[3]}")
    else:
        print("No events found matching the search term.")
def send_invitations():
    """Sends invitations for an event by simulating email output to the console."""
    event_name = input("Enter the Event Name to send invitations for: ")

# Read attendee details from the file
    with open(ATTENDEE_FILE, 'r') as file:
        reader = csv.reader(file)
        attendees = list(reader)
# Filter attendees for the specified event
    event_attendees = [attendee for attendee in attendees[1:] if attendee[0].lower() == event_name.lower()]

    if event_attendees:
        print(f"\nSending invitations for '{event_name}'...\n")
        for attendee in event_attendees:
            name, email, rsvp_status = attendee[1], attendee[2], attendee[3]
            print(f"To: {email}")
            print(f"Subject: Invitation to {event_name}")
            print(f"Hi {name},\n")
            print(f"You are invited to the event '{event_name}'. Please let us know your RSVP status.\n")
            print("-" * 40)
        print("\nInvitations sent successfully!")
    else:
        print(f"No attendees found for the event '{event_name}'.")

#Additional Features(Display Calander)
def display_calendar():
    """Displays a monthly calendar with event dates highlighted."""
# Get the current year and month
    today = datetime.now()
    year = today.year
    month = today.month

# Read event dates from the events file
    with open(EVENT_FILE, 'r') as file:
        reader = csv.reader(file)
        events = list(reader)

# Extract dates of events for the current month and year
    event_dates = []
    for event in events[1:]:
        event_date = datetime.strptime(event[1], "%Y-%m-%d %H:%M").date()
        if event_date.year == year and event_date.month == month:
            event_dates.append(event_date.day)

# Generate the calendar
    cal = calendar.TextCalendar()
    cal_str = cal.formatmonth(year, month)

# Highlight event dates
    for day in event_dates:
        cal_str = cal_str.replace(f"{day:2d}", f"*{day:2d}" if day >= 10 else f"*{day}")

# Display the calendar
    print(f"Calendar for {calendar.month_name[month]} {year}:\n")
    print(cal_str)
    print("* indicates dates with scheduled events.")

#Additional Features (Export Data)

# Function to export event and attendee data to a CSV file
def export_data():
    """Exports event and attendee data to a single CSV file."""
    export_file = "exported_event_data.csv"

    with open(EVENT_FILE, 'r') as event_file, open(ATTENDEE_FILE, 'r') as attendee_file:
# Read events and attendees
        event_reader = csv.reader(event_file)
        attendee_reader = csv.reader(attendee_file)

        events = list(event_reader)
        attendees = list(attendee_reader)

# Write to the export file
    with open(export_file, 'w', newline='') as file:
        writer = csv.writer(file)

# Write event data
        writer.writerow(["Event Data"])
        writer.writerow(["Event Name", "Date and Time", "Location", "Description"])
        for event in events[1:]:
            writer.writerow(event)

        writer.writerow([])  # Add a blank row for separation

# Write attendee data
        writer.writerow(["Attendee Data"])
        writer.writerow(["Event Name", "Attendee Name", "Attendee Email", "RSVP Status"])
        for attendee in attendees[1:]:
            writer.writerow(attendee)

    print(f"Event and attendee data exported successfully to {export_file}.")
```
